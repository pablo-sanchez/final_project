## Time to make some groupby

* Average and median salary of each type of job position and maybe regarding the company size
* Average and median salary by sector and maybe the zone
* Average and median salary depending on the skill they have

```python
job_median=df.pivot_table(index=['job_simpl'], columns=['seniority'], values=['salary_estimate'], aggfunc=['median'])
job_median.columns = ['Senior','junior','mid']
job_median.reset_index()
job_median
```

```python
job_mean=df.pivot_table(index=['seniority'], columns=['job_simpl'], values=['salary_estimate'], aggfunc=['mean'])
job_mean.columns = ['data analyst','data engineer','data scientist','machine learning engineer']
job_mean
```

```python
fig, ax = plt.subplots(2,2, figsize=(12,15))
c=['green','blue','purple']
fig.suptitle('Average salary by data role and experience', fontsize="x-large")

x1 = ['Junior Data Analyst', 'Mid Data Analyst', 'Senior Data Analyst']
y1= [job_mean['data analyst'][1],job_mean['data analyst'][2],job_mean['data analyst'][0]]
ax[0,0].bar(x1,y1,color=c)
ax[0,0].set_ylim(0,135000)
ax[0,0].set_ylabel('Average salary ($USD)')

x2 = ['Junior Data Engineer', 'Mid Data Engineer', 'Senior Data Engineer']
y2= [job_mean['data engineer'][1],job_mean['data engineer'][2],job_mean['data engineer'][0]]
ax[0,1].bar(x2,y2,color=c)
ax[0,1].set_ylim(0,135000)
ax[0,1].set_ylabel('Average salary ($USD)')

x3 = ['Junior Data Scientist', 'Mid Data Scientist', 'Senior Data Scientist']
y3= [job_mean['data scientist'][1],job_mean['data scientist'][2],job_mean['data scientist'][0]]
ax[1,0].bar(x3,y3,color=c)
ax[1,0].set_ylim(0,135000)
ax[1,0].set_ylabel('Average salary ($USD)')

x4 = ['Junior ML Engineer', 'Mid ML Engineer', 'Senior ML Engineer']
y4= [job_mean['machine learning engineer'][1],job_mean['machine learning engineer'][2],job_mean['machine learning engineer'][0]]
ax[1,1].bar(x4,y4,color=c)
ax[1,1].set_ylim(0,135000)
ax[1,1].set_ylabel('Average salary ($USD)')

plt.savefig(config['images']+'role_and_seniority_trans.png', dpi=300, transparent=True)
plt.savefig(config['images']+'role_and_seniority.png', dpi=300)
plt.show()
```

```python
size = df.pivot_table(index=['size'], values=['salary_estimate'], aggfunc=['mean'])
size=size.reset_index()
size.columns = ['size','average_salary']
size=size.sort_values(by=['average_salary'], ascending=False)
```

```python
c1=['red','orange','green','blue','purple']
size.plot(x='size',y='average_salary', kind='bar', color=c1, legend=False)
plt.ylabel('Average salary ($USD)')

plt.savefig(config['images']+'salary_and_size_trans.png', dpi=300, transparent=True)
plt.savefig(config['images']+'salary_and_size.png', dpi=300)
plt.show()
```

**The bigger the company the higher salary, however there is not such difference once the company reaches a size of at least 500 employees, in fact these companies pays more in comparison with companies over 10.000 employees.**

```python
sector=df.pivot_table(index=['sector'], values=['salary_estimate'], aggfunc=['mean'])
sector=sector.reset_index()
sector.columns=['sector','average_salary']
sector=sector.sort_values(by=['average_salary'], ascending=False)
```

```python
c2=['cyan','gold','limegreen','navy','gray','darkorange','aliceblue','indigo']
sector.plot(x='sector',y='average_salary', kind='bar', color=c1, legend=False)
plt.ylabel('Average salary ($USD)')

plt.savefig(config['images']+'salary_and_sector_trans.png', dpi=300, transparent=True)
plt.savefig(config['images']+'salary_and_sector.png', dpi=300)
plt.show()
```

**We can see that IT and telecomunications is the sector that pays the most**
